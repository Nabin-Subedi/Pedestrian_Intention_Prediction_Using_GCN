```python
def get_data(file_path):
    mytree=et.parse(file_path)
    myroot=mytree.getroot()
    data={}
    person_number=0
    for track in myroot.findall("./track[@label='pedestrian']"):
        frame_dict={}
        for i,box in enumerate(track):
            if i==0:
                #person_id=box[0].text
                person_id='person_'+str(person_number)
                person_number+=1
                #print(box[0].text)
            frame_no=int(box.attrib['frame'])
            bbox=[box.attrib['xtl'],box.attrib['ytl'],box.attrib['xbr'],box.attrib['ybr']]
            c_nc=box[5].text
    #         print(box.attrib['frame'],box.attrib['xtl'],box.attrib['ytl'],box.attrib['xbr'],box.attrib['ybr'])
    #         print(box[5].text)
            frame_dict[frame_no]={'bbox':bbox,'c_nc':c_nc}
        data[person_id]=frame_dict
        #print(track.attrib)
    return data
```

```python
#for reading json file from JAAD and drawing 
def get_bbox(frame_num):
    video_num='0000'+str(1)
    video_num=video_num[-4:]
    json_path='E:/wuhan/lab/Jaad/nbn_annot_json/video_'+video_num+'.json'
    bboxes=[]
    c_ncs=[]
    with open(json_path,'r') as f:
        data=json.load(f)
    for i in range(len(data)):
        try:
            bbox=data['person_'+str(i)][str(frame_num)]['bbox']
            c_nc=data['person_'+str(i)][str(frame_num)]['c_nc']
        except:
            continue
        bboxes.append(bbox)
        c_ncs.append(c_nc)
    return bboxes,c_ncs
```

```python
cap = cv2.VideoCapture(video_path)
if cap.isOpened() is False:
    print("Error opening video stream or file")
fpss = cap.get(cv2.CAP_PROP_FPS)
print(fpss)
```

```python
writeVideo_flag=False
frame_num=0
while True:
    ret_val, image = cap.read()
    if not ret_val:
        break
    bboxes,c_ncs=get_bbox(frame_num)
    for bbox,c_nc in zip(bboxes,c_ncs):
        cv2.rectangle(image, (int(float(bbox[0])), int(float(bbox[1]))), (int(float(bbox[2])), int(float(bbox[3]))), (255, 0, 0), 2)
        cv2.putText(image,str(c_nc)  , (int(float(bbox[0])), int(float(bbox[1]))), 0, 1e-3 * image.shape[0] , (255, 255, 0), 2)
    
    cv2.imshow('result', image)
    
    ### For Saving video ###
    if writeVideo_flag:
        if frame_num == 0:   # It's use to intialize video writer ones
            out = cv2.VideoWriter(file_write_name+'_output.avi',cv2.VideoWriter_fourcc('M', 'J', 'P', 'G'),
                    20,(image.shape[1],image.shape[0]))
        out.write(image)
    frame_num+=1
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
if writeVideo_flag==True:
    out.release()
cv2.destroyAllWindows()
print("The video was successfully saved")
```
