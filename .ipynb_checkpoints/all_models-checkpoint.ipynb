```python
def get_einsum(x,temporal=300):
  A_stack = np.stack([A for _ in range(temporal)])
  A_stack = tf.convert_to_tensor(A_stack,dtype='float')
  x=tf.transpose(x,perm=[0,3,1,2])
  x=tf.einsum('nctv,tvw->nctw', x, A_stack)
  x=tf.transpose(x,perm=[0,2,3,1])
  return x

def change_shape(x):
  x=tf.transpose(x,perm=[0,1,3,2])
  return x


i=Input(shape=(300,25,2))

x=Lambda(get_einsum , arguments={'temporal':300})(i)
x=Conv2D(filters=64,kernel_size=(3,3),padding='same')(x)
x=LeakyReLU()(x)
# x=Activation('relu')(x)
x=Dropout(0.5)(x)

x=Lambda(get_einsum , arguments={'temporal':300})(x)
x=Conv2D(filters=32,kernel_size=(3,3),padding='same')(x)
x=LeakyReLU()(x)
# x=Activation('relu')(x)
x=Dropout(0.5)(x)

x=Lambda(get_einsum , arguments={'temporal':300})(x)
x=Conv2D(filters=16,kernel_size=(5,5),padding='same')(x)
x=LeakyReLU()(x)
x=Dropout(0.2)(x)

x=Lambda(change_shape)(x)
x=GlobalAveragePooling2D()(x)

x=Dense(units=10)(x)
x=LeakyReLU()(x)
# x=Activation('relu')(x)
x=Dropout(0.2)(x)
x=Dense(units=1)(x)
x=Activation('sigmoid')(x)

gcn_model=Model(inputs=i,outputs=x)
```

```python
def iou_end(bbox, candidates):
    """Computes intersection over union.

    Parameters
    ----------
    bbox : ndarray
        A bounding box in format `(top left x, top left y, width, height)`.
    candidates : ndarray
        A matrix of candidate bounding boxes (one per row) in the same format
        as `bbox`.

    Returns
    -------
    ndarray
        The intersection over union in [0, 1] between the `bbox` and each
        candidate. A higher score means a larger fraction of the `bbox` is
        occluded by the candidate.

    """
    bbox_tl, bbox_br = bbox[:2], bbox[:2] + bbox[2:]
    candidates_tl = candidates[:, :2]
    candidates_br = candidates[:, :2] + candidates[:, 2:]

    tl = np.c_[np.maximum(bbox_tl[0], candidates_tl[:, 0])[:, np.newaxis],
               np.maximum(bbox_tl[1], candidates_tl[:, 1])[:, np.newaxis]]
    br = np.c_[np.minimum(bbox_br[0], candidates_br[:, 0])[:, np.newaxis],
               np.minimum(bbox_br[1], candidates_br[:, 1])[:, np.newaxis]]
    wh = np.maximum(0., br - tl)

    area_intersection = wh.prod(axis=1)
    area_bbox = bbox[2:].prod()
    area_candidates = candidates[:, 2:].prod(axis=1)
    return area_intersection / (area_bbox + area_candidates - area_intersection)
```

```python
def calculate_boxes(keypoints):
    keypoints=keypoints[np.invert(keypoints[:,2]==0)]
    #keypoints=keypoints[np.all(keypoints != 0, axis=1)] # for comparing if all three values in a row is equal to zero. 
    x1=keypoints[:,0].min()
    y1=keypoints[:,1].min()
    x2=keypoints[:,0].max()
    y2=keypoints[:,1].max()
    width=x2-x1
    height=y2-y1
    height=height+(0.08*height)
    y1=max((y1-(0.08*height)),0)
    bbox=[x1,y1,width,height]
    return bbox

def get_head_and_boxes(file_name):
    bboxes=[]
    all_keypoints=[]
    with open (file_name) as f:
        data= json .load(f)
    for i in range(len(data['people'])):
        keypoints=data['people'][i]['pose_keypoints_2d']
        keypoints=np.array(keypoints)
        keypoints=keypoints.reshape(25,3)
        keypoints_array=keypoints.copy()
        keypoints=keypoints[:,:2]
        keypoints=keypoints.tolist()
        bbox=calculate_boxes(keypoints_array)
        #bbox=[keypoints[0][0]-10,keypoints[0][1]-10,25,25]
        bboxes.append(bbox)
        all_keypoints.append(keypoints)
#     print(f'number of poeple: {len(data["people"])}')
#     print(bboxes)
    return bboxes,all_keypoints
```

```python
def get_keypoints_per_person(bbox,boxes,all_keypoints):
    '''
    Comparing tracked boxes with original boxes based on iou 
    and getting head co-ordinate of that person in paricular bounding box.
    Note: head_cord in'head_cords' for each person is in same position as bounding box in 'boxes'  
    '''
    iou_values=iou_end(bbox, boxes)
    iou_dict={'position':np.argmax(iou_values),'iou_value':np.max(iou_values)}
    keypoints_single_person=all_keypoints[iou_dict['position']]
    return keypoints_single_person
```

```python
def store_data_for_gcn(data_for_gcn,person_id,frame_num,person_keypoints):
    ''' Updates data_for_fall. Stores new data and pops old data.
        Stores 5 data for each id.
    '''
    if person_id in data_for_gcn.keys():
        data_for_gcn[person_id].append(person_keypoints)
        if len(data_for_gcn[person_id])>300:
            data_for_gcn[person_id].pop(0)
    else:
        data_for_gcn[person_id]=[person_keypoints]
```

```python
def get_intention(data_for_gcn,id_for_intention,gcn_model):
    X_prep=[]
    id_intent={}
    id_to_predict=[]
    for person_id in id_for_intention:
        if len(data_for_gcn[person_id])>100:
            id_to_predict.append(person_id)
            x_person=data_for_gcn[person_id]
            for _ in range(300-len(data_for_gcn[person_id])):
                x_person.append(np.zeros((25,2)).tolist())
            x_person_avg=get_average(x_person)
            X_prep.append(x_person_avg)
    if len(X_prep)==0:
        return id_intent
    X=np.array(X_prep)
    X[:, :, :, 0] /= 1920.0
    X[:, :, :, 1] /= 1080.0
    prediction=gcn_model.predict(X)
    prediction=prediction>0.5
    for per_id,pred in zip(id_to_predict,prediction):
        if pred==1:
            intent='Crossing'
        elif pred==0:
            intent='Not-crossing'
        id_intent[per_id]=intent
        
    return id_intent

def get_average(x_person):
    person_array=np.array(x_person)
    for frame in range(300):
        array_per_frame=person_array[frame]
        if np.all(array_per_frame==0):
            continue
        if np.all(array_per_frame!=0):
            continue
        if np.all(array_per_frame[:,0] !=0) or np.all(array_per_frame[:,1] !=0):
            continue
        col_1_sum=np.sum(array_per_frame[:,0])
        col_2_sum=np.sum(array_per_frame[:,1])
        total_non_zero_element=np.sum(array_per_frame[:,0] != 0)
        col_1_avg=col_1_sum/total_non_zero_element
        col_2_avg=col_2_sum/total_non_zero_element
        person_array[frame][:,0]=(person_array[frame][:,0]==0)*col_1_avg+person_array[frame][:,0]
        person_array[frame][:,1]=(person_array[frame][:,1]==0)*col_2_avg+person_array[frame][:,1]
    return person_array.tolist()
```

```python
for frame in range(0,280,10):
#     color = list(np.random.random(size=3) * 256)
    color=colors[np.random.randint(17)]
    #color=[0,128,0]
    for i in range(25):
        cv2.circle(img,tuple(all_keys[frame][i].astype(int)*3),5,[255, 85, 0],-1)
    for pair in pairs:
        if all_keys[frame][pair[0]][0] ==0 or all_keys[frame][pair[1]][0]==0:
            continue
        cv2.line(img,tuple(all_keys[frame][pair[0]].astype(int)*3),tuple(all_keys[frame][pair[1]].astype(int)*3),color,5,1)
```
